```python
# get cities info
def City():
    cities = []
    f = open("/kaggle/input/dataa/280.txt")
    for i in f.readlines():
        node_city_val = i.split()
        cities.append(
            [node_city_val[0], float(node_city_val[1]), float(node_city_val[2])]
        )

    return cities
City()
```

```python
cities = {item[0]: tuple(item[1:]) for item in City()}
```

```python
import random
import matplotlib.pyplot as plt


population_size = 2000
mutation_rate = 0.01
num_generations = 1500

def create_initial_population():
    population = []
    for _ in range(population_size):
        cities_list = list(cities.keys())
        random.shuffle(cities_list)
        population.append(cities_list)
    return population

def calculate_fitness(individual):
    total_distance = 0
    for i in range(len(individual) - 1):
        city_a = individual[i]
        city_b = individual[i + 1]
        distance = ((cities[city_b][0] - cities[city_a][0])**2 + (cities[city_b][1] - cities[city_a][1])**2)**0.5
        total_distance += distance
    total_distance += ((cities[individual[-1]][0] - cities[individual[0]][0])**2 + (cities[individual[-1]][1] - cities[individual[0]][1])**2)**0.5
    return 1 / total_distance

def crossover(parent1, parent2):
    cut_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:cut_point] + [city for city in parent2 if city not in parent1[:cut_point]]
    child2 = parent2[:cut_point] + [city for city in parent1 if city not in parent2[:cut_point]]
    return child1, child2

def mutate(individual):
    idx1 = random.randint(0, len(individual) - 1)
    idx2 = random.randint(0, len(individual) - 1)
    individual[idx1], individual[idx2] = individual[idx2], individual[idx1]
    return individual

def select_new_population(population):
    fitness_scores = [calculate_fitness(individual) for individual in population]
    total_fitness = sum(fitness_scores)
    probabilities = [score / total_fitness for score in fitness_scores]
    new_population = random.choices(population, probabilities, k=population_size)
    return new_population

def calculate_total_distance(route):
    total_distance = 0
    for i in range(len(route) - 1):
        city_a = route[i]
        city_b = route[i + 1]
        distance = ((cities[city_b][0] - cities[city_a][0])**2 + (cities[city_b][1] - cities[city_a][1])**2)**0.5
        total_distance += distance
    total_distance += ((cities[route[-1]][0] - cities[route[0]][0])**2 + (cities[route[-1]][1] - cities[route[0]][1])**2)**0.5
    return total_distance

def plot_tsp_route(route):
    x = [cities[city][0] for city in route]
    y = [cities[city][1] for city in route]
    x.append(x[0])
    y.append(y[0])

    plt.plot(x, y, 'ro-')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('TSP Route')
    plt.grid(True)
    plt.show()

def genetic_algorithm():
    population = create_initial_population()

    for _ in range(num_generations):
        new_population = []

        while len(new_population) < population_size:
            parent1 = random.choice(population)
            parent2 = random.choice(population)
            child1, child2 = crossover(parent1, parent2)

            if random.random() < mutation_rate:
                child1 = mutate(child1)
            if random.random() < mutation_rate:
                child2 = mutate(child2)

            new_population.append(child1)
            new_population.append(child2)

        population = select_new_population(new_population)

    final_best_individual = max(population, key=calculate_fitness)
    final_best_fitness = calculate_fitness(final_best_individual)
    final_best_route = ' -> '.join(final_best_individual + [final_best_individual[0]])

    total_distance = calculate_total_distance(final_best_individual)

    print("\nFinal result:")
    print(f"Best route: {final_best_route}")
    print(f"Best fitness: {final_best_fitness}")
    print(f"Total distance: {total_distance}")

    plot_tsp_route(final_best_individual)

genetic_algorithm()
```
